7. Giá trị bị mất 

Vấn đề 7: “Missing values”, vấn đề này xảy ra tại các cột “Age”, “Weight” và “Heart 
Rate”. Thiếu dữ liệu (dữ liệu không đầy đủ) là vấn đề xảy ra nhiều trong các nguồn dữ 
liệu do nhiều nguyên nhân chủ quan lẫn khách quan.Có một vài giải pháp để xử lý vấn đề 
này, chủ yếu dựa trên kinh nghiệm và nghiệp vụ về tập dữ liệu đó. Một số giải pháp đưa 
đề xuất từ chuyên gia như sau: 
a. Deletion: Remove records with missing values 
b. Dummy substitution: Replace missing values with a dummy but valid value: 
e.g.: 0 for numerical values. 
c. Mean substitution: Replace the missing values with the mean. 
d. Frequent substitution: Replace the missing values with the most frequent item. 
e. Improve the data collector: Your business folk will talk to the clients and inform 
them about why it is worth fixing the problem with the data collector. 
…… 

a. Deletion: Remove records with missing values 

```python
df_a = df.copy()

df_a.dropna(subset=["Age", "Weight"], how="any", inplace=True)

print("--- Kết quả sau khi XÓA các dòng bị khuyết ---")
df_a
```

b. Dummy substitution: Replace missing values with a dummy but valid value: 
e.g.: 0 for numerical values. 

```python
# Tạo bản sao để thử nghiệm phương pháp b
df_b = df.copy()

df_b["Age"] = df_b["Age"].fillna(0)
df_b["Weight"] = df_b["Weight"].fillna(0)

print("--- Kết quả sau khi thay thế giá trị khuyết bằng số 0 ---")
df_b
```

c. Mean substitution: Replace the missing values with the mean. 

```python
df_c = df.copy()

if df_c["Weight"].dtype == "object":
    df_c["Weight"] = df_c["Weight"].str.replace("kgs", "").astype(float)

mean_age = df_c["Age"].mean()
mean_weight = df_c["Weight"].mean()

# Điền giá trị trung bình vào các ô trống (làm tròn để số đẹp hơn)
df_c["Age"] = df_c["Age"].fillna(round(mean_age, 1))
df_c["Weight"] = df_c["Weight"].fillna(round(mean_weight, 1))

print("--- Kết quả sau khi thay thế bằng giá trị TRUNG BÌNH ---")
df_c
```

d. Frequent substitution: Replace the missing values with the most frequent item.

```python
# Tạo bản sao để thử nghiệm phương pháp d
df_d = df.copy()

mode_age = df_d["Age"].mode()[0]
mode_weight = df_d["Weight"].mode()[0]

# Điền giá trị xuất hiện nhiều nhất vào các ô trống
df_d["Age"] = df_d["Age"].fillna(mode_age)
df_d["Weight"] = df_d["Weight"].fillna(mode_weight)

print("--- Kết quả sau khi thay thế bằng giá trị XUẤT HIỆN NHIỀU NHẤT ---")
df_d
```

e. Improve the data collector: Your business folk will talk to the clients and inform 
them about why it is worth fixing the problem with the data collector. 

```python
# Tạo bản sao để thử nghiệm phương pháp d
df_d = df.copy()

mode_age = df_d["Age"].mode()[0]
mode_weight = df_d["Weight"].mode()[0]

df_d["Age"] = df_d["Age"].fillna(mode_age)
df_d["Weight"] = df_d["Weight"].fillna(mode_weight)

print("--- Kết quả sau khi thay thế bằng giá trị XUẤT HIỆN NHIỀU NHẤT ---")
df_d
```

 Vấn đề 8: “một cột chứa quá nhiều thông tin cần được phân rã”, như trong bài toán này ta 
thấy header “m0006” chứa các nội dung bao gồm: m → male, 1218 ~ 12-18 (mm-dd). 
Còn giá trị thì là kết quả huyết áp. 

```python
import numpy as np
import pandas as pd

# 1. ĐỌC LẠI FILE VÀ ĐỒNG BỘ NHANH (Bắt buộc để reset dữ liệu trong bộ nhớ Jupyter, tránh lỗi chạy lại bị NaN)
column_names = [
    "Id",
    "Name",
    "Age",
    "Weight",
    "m0006",
    "m0612",
    "m1218",
    "f0006",
    "f0612",
    "f1218",
]
df = pd.read_csv("patient_heart_rate.csv", names=column_names, on_bad_lines="skip")
df.dropna(how="all", inplace=True)

# Đồng bộ lại các bước tiền xử lý trước đó
df[["Firstname", "Lastname"]] = df["Name"].astype(str).str.split(expand=True)
df = df.drop("Name", axis=1)

for i in range(0, len(df["Weight"])):
    x = str(df["Weight"].iloc[i])
    if "lbs" in x:
        df.at[df.index[i], "Weight"] = str(int(float(x[:-3]) / 2.2)) + "kgs"


# 2. TIẾN HÀNH MELT BẢNG DỮ LIỆU
df = pd.melt(
    df,
    id_vars=["Id", "Age", "Weight", "Firstname", "Lastname"],
    value_name="PulseRate",
    var_name="sex_and_time",
)

# Sắp xếp để gom nhóm dữ liệu theo từng người cho giống ảnh mẫu
df = df.sort_values(by=["Id", "Age", "Weight", "Firstname", "Lastname"])


# 3. SỬ DỤNG REGEX ĐỂ TRÍCH XUẤT CÁC THÔNG TIN
# ^(\D): Lấy ký tự chữ đầu tiên (m hoặc f)
# (\d{2})(\d{2}): Tách 4 chữ số phía sau thành 2 nhóm giờ
extract_df = df["sex_and_time"].str.extract(r"^(\D)(\d{2})(\d{2})")
extract_df.columns = ["Sex", "hours_lower", "hours_upper"]
extract_df.index = df.index  # Đồng bộ chỉ số hàng để tuyệt đối không bị lỗi NaN khi gán


# 4. GÁN DỮ LIỆU VÀ CHUYỂN ĐỔI THÀNH CHỮ 'M' / 'F' IN HOA
df["Sex"] = extract_df["Sex"].str.upper()  # .upper() sẽ biến 'm'->'M' và 'f'->'F'
df["Time"] = extract_df["hours_lower"] + "-" + extract_df["hours_upper"]


# 5. DỌN DẸP BẢNG DỮ LIỆU
df = df.drop("sex_and_time", axis=1)
df["PulseRate"] = df["PulseRate"].replace("-", np.nan)

# Sắp xếp lại thứ tự các cột chuẩn theo yêu cầu của bạn
ordered_cols = [
    "Id",
    "Age",
    "Weight",
    "Firstname",
    "Lastname",
    "PulseRate",
    "Sex",
    "Time",
]
df = df[ordered_cols]

print("--- Bảng dữ liệu sạch chuẩn cho Vấn đề 8 (Cột Sex dạng M/F) ---")
df
```

11. Hãy khảo sát tỉ lệ dữ liệu thiếu trên biến huyết áp. Dữ liệu bị thiếu thì hãy xử lý bằng 
phương pháp sau 
• Thay thế bằng giá trị trung bình liền trước và liền sau của người đó. Nếu không 
được thì dùng 2) 
• Thay thế bằng giá trị trung bình 2 giá liền trước của người đó. Nếu không được 
thì dùng 3) 
• Thay thế bằng giá trị trung bình 2 giá liền sau của người đó. Nếu không được thì 
dùng 4) 
• Trung bình của các giá trị huyết áp của người đó. Nếu không được thì dùng 5). 
• Trung bình của các giá trị huyết áp của nhóm giới tính. Nếu không được thì dùng 
6) 
• Trung bình của các
giá trị dữ liệu. Nếu không được thì thay bằng mức ổn định 
trong y học.   

```python

df["PulseRate"] = pd.to_numeric(df["PulseRate"].replace("-", np.nan), errors="coerce")

# 1. Khảo sát tỉ lệ dữ liệu thiếu sau khi đồng bộ
missing_ratio = df["PulseRate"].isna().mean() * 100
print(f"Tỉ lệ dữ liệu khuyết trên cột PulseRate: {missing_ratio:.2f}%\n")


# 2. Định nghĩa hàm xử lý nội suy y tế theo các tầng điều kiện từ chuyên gia
def fill_pulse_rate(group):
    # Lấy mảng giá trị của người đó
    pulse = group["PulseRate"].copy()

    # Tầng 1: Trung bình của liền trước và liền sau
    t1 = (pulse.shift(1) + pulse.shift(-1)) / 2
    pulse = pulse.fillna(t1)

    # Tầng 2: Trung bình của 2 giá trị liền trước
    t2 = (pulse.shift(1) + pulse.shift(2)) / 2
    pulse = pulse.fillna(t2)

    # Tầng 3: Trung bình của 2 giá trị liền sau
    t3 = (pulse.shift(-1) + pulse.shift(-2)) / 2
    pulse = pulse.fillna(t3)

    # Tầng 4: Trung bình của các giá trị hiện có của chính người đó
    t4 = pulse.mean()
    pulse = pulse.fillna(t4)

    group["PulseRate"] = pulse
    return group


# Áp dụng 4 tầng đầu tiên cho từng cá nhân (group theo Id)
df = df.groupby("Id", group_keys=False).apply(fill_pulse_rate)

# Tầng 5: Nếu vẫn khuyết, thay bằng trung bình của nhóm giới tính (Sex)
df["PulseRate"] = df.groupby("Sex")["PulseRate"].transform(
    lambda x: x.fillna(x.mean())
)

# Tầng 6: Nếu vẫn khuyết, thay bằng mức ổn định mặc định trong y học (70 bpm)
df["PulseRate"] = df["PulseRate"].fillna(70)

# Làm tròn chỉ số huyết áp về số nguyên cho sạch đẹp
df["PulseRate"] = df["PulseRate"].round().astype(int)

print("--- Hoàn thành xử lý nội suy khuyết dữ liệu PulseRate ---")
df.head(10)
```
