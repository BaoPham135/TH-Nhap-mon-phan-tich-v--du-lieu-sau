1.  Liệt kê các tên của corpus (kho ngữ liệu) có sẵn trong NLTK

```python
import os
import nltk

# Tìm đường dẫn đến thư mục chứa corpora của NLTK
corpora_path = nltk.data.find('corpora')

print("Danh sách các corpus phổ biến có trong thư mục dữ liệu của NLTK:")

# Sử dụng os.listdir để liệt kê các thư mục/file con bên trong
for name in os.listdir(corpora_path):
    # Kiểm tra nếu là thư mục hoặc các file dữ liệu (bỏ qua các file hệ thống ẩn nếu có)
    if not name.startswith('.'):
        print("- ", name)
```

4. Loại bỏ các stopwords từ một văn bản cho trước

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Natural Language Processing with Python and NLTK is very interesting and useful."
stop_words = set(stopwords.words('english'))

# Tách từ (Tokenize)
words = word_tokenize(text)

# Lọc bỏ stopwords
filtered_text = [w for w in words if w.lower() not in stop_words]

print("Văn bản gốc:", words)
print("Sau khi lọc stopwords:", filtered_text)
```

5. Bỏ qua (loại trừ) một số từ ra khỏi danh sách stopwords gốc

```python
from nltk.corpus import stopwords

# Lấy danh sách gốc
stop_words = set(stopwords.words('english'))

# Các từ muốn giữ lại
words_to_keep = {'not', 'no'}
filtered_stopwords = stop_words - words_to_keep

print("Từ 'not' có trong danh sách gốc không?", 'not' in stop_words)
print("Từ 'not' có trong danh sách mới không?", 'not' in filtered_stopwords)
```

6. Tìm định nghĩa và ví dụ của một từ bằng WordNet

```python
from nltk.corpus import wordnet

word = "bank"
synsets = wordnet.synsets(word)

print(f"Từ '{word}' có {len(synsets)} nghĩa (synsets) khác nhau trong WordNet.\n")
for idx, syn in enumerate(synsets[:3], 1): # Lấy tối đa 3 nghĩa đầu tiên
    print(f"Nghĩa {idx} ({syn.name()}):")
    print(" - Định nghĩa:", syn.definition())
    print(" - Ví dụ:", syn.examples())
```

7. Tìm tập hợp từ đồng nghĩa (Synonyms) và trái nghĩa (Antonyms)

```python
from nltk.corpus import wordnet

synonyms = []
antonyms = []

word = "good"
for syn in wordnet.synsets(word):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(f"Từ đồng nghĩa của '{word}':", set(synonyms))
print(f"Từ trái nghĩa của '{word}':", set(antonyms))
```

9. So sánh sự giống nhau của hai danh từ

```python
from nltk.corpus import wordnet

# Lấy nghĩa danh từ (noun) của hai từ
dog = wordnet.synset('dog.n.01')
cat = wordnet.synset('cat.n.01')

# Tính độ tương đồng Wu-Palmer
similarity = dog.wup_similarity(cat)
print(f"Độ tương đồng giữa 'dog' và 'cat': {similarity:.4f}")
```

10. So sánh sự giống nhau của hai động từ

```python
from nltk.corpus import wordnet

# Lấy nghĩa động từ (verb) của hai từ
run = wordnet.synset('run.v.01')
walk = wordnet.synset('walk.v.01')

similarity = run.wup_similarity(walk)
print(f"Độ tương đồng giữa 'run' và 'walk': {similarity:.4f}")
```

12. Trộn và in 15 kết hợp ngẫu nhiên tên nam và nữ có gắn nhãn

```python
import random
from nltk.corpus import names

# Gắn nhãn cho từng tên
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Trộn ngẫu nhiên danh sách
random.shuffle(labeled_names)

print("15 kết hợp ngẫu nhiên đầu tiên được gắn nhãn:")
for name, gender in labeled_names[:15]:
    print(f"- Tên: {name} ({gender})")
```

13. Trích xuất ký tự cuối cùng của tên và tạo mảng dữ liệu mới

```python
from nltk.corpus import names

labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Tạo mảng mới chứa: (Kí tự cuối cùng, Nhãn giới tính)
last_letters_dataset = [(name[-1].lower(), gender) for name, gender in labeled_names]

print("Mẫu 10 phần tử đầu tiên trong mảng mới:")
for char, label in last_letters_dataset[:10]:
    print(f"Chữ cái cuối: '{char}' -> Nhãn: {label}")
```
